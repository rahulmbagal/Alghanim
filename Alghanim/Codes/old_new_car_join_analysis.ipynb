```python
data_USED_CAR = pd.read_excel("used_cars_cleaned.xlsx")
data_NEW_CAR = pd.read_excel("new_cars_cleaned.xlsx")
print (data_USED_CAR.shape)
print (data_NEW_CAR.shape)
```

```python
pd.merge(data_USED_CAR, data_NEW_CAR, on = "CUSTOMERID", how = "inner")
```

```python
concat_data = pd.concat([data_USED_CAR, data_NEW_CAR])
concat_data.head(2)
```

```python
temp_1 = concat_data.copy(deep= True)
temp_1 = temp_1[['MODEL', 'TRANSACTION_DATE_1']].sort_values(['MODEL', 'TRANSACTION_DATE_1'], ascending=[True, True])
temp_1 = temp_1.drop_duplicates(keep='first')
df = temp_1.copy()


df['diff']=df.groupby('MODEL')['TRANSACTION_DATE_1'].diff() / np.timedelta64(1, 'D')
df['diff'] = df['diff'].fillna(0)
df['diff'] = df['diff'].astype(int)

df[df["diff"] > 0].groupby('MODEL').agg({'diff': "mean"})
```

```python
temp_1 = concat_data.copy(deep= True)
temp_1 = temp_1[['GENDER', 'TRANSACTION_DATE_1']].sort_values(['GENDER', 'TRANSACTION_DATE_1'], ascending=[True, True])
temp_1 = temp_1.drop_duplicates(keep='first')
df = temp_1.copy()

df['GENDER'] = np.where(df['GENDER']=='n?a', 'Not_Available', df['GENDER'])
df.fillna({'GENDER' : 'Not_Available'}, inplace = True)
df['diff']=df.groupby('GENDER')['TRANSACTION_DATE_1'].diff() / np.timedelta64(1, 'D')
df['diff'] = df['diff'].fillna(0)
df['diff'] = df['diff'].astype(int)

df[df["diff"] > 0].groupby('GENDER').agg({'diff': "mean"})
```
